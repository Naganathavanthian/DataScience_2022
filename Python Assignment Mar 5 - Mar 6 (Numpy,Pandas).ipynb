5.	Create two arrays of any dimension and perform vstack and hstack
6.	Append two arrays using axis = 0 and axis = 1

We have combined both the tasks in the following example

```python
array_a=np.full((3,3),5)
```

```python
array_b=np.full((3,3),10)
```

```python
# vstack stacks the data in the array in the rows, vertically .so it is called as vstack(vertical stack)
np.vstack((array_a,array_b))
```

```python
# vstack is equivalent to appending the array in the axis=0
np.append(array_a,array_b,axis=0)
```

```python
# hstack stacks the data in the array in the columns, horizontally .so it is called as hstack(horizontal stack)
np.hstack((array_a,array_b))
```

```python
# hstack is equivalent to appending the array in the axis=1
np.append(array_a,array_b,axis=1)
```

14.	Split the below table into half and perform merge (inner, left and right) on Region. One example for each type of merge (inner, left and right)

```python
df6=df4[9:]
```

```python
df6
```

```python
pd.merge(df5,df6,on='U.S. Region')
```

```python
df5.merge(df6,on='U.S. Region',how='left')
```

```python
df5.merge(df6,on='U.S. Region',how='right')
```

```python
df5.merge(df6,on='U.S. Region',how='inner')
```
