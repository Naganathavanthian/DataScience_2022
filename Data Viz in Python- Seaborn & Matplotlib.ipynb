```python
flight_df=pd.read_csv("flight_delays.csv",index_col='Month')
```

```python
flight_df
```

```python
plt.figure(figsize=(15,10))
plt.title("Avg Arrival delays for airline WN by Month")
sns.barplot(x=flight_df.index,y=flight_df['WN'],data=flight_df)

# add the labels
plt.xlabel(" Month ")
plt.xlabel(" Arrival_delays ")
```

#### Heatmap

```python

plt.figure(figsize=(15,10))
plt.title("Avg Arrival delays for airline WN by Month")

#heatmap
sns.heatmap(data=flight_df,annot=True,cmap='YlGnBu')

```

```python

plt.figure(figsize=(15,10))
plt.title("Avg Arrival delays for airline WN by Month")

#heatmap
corr=flight_df.corr()
sns.heatmap(data=corr,annot=True,cmap='YlGnBu')

```
