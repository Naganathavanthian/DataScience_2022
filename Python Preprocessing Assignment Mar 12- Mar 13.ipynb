12.	Read the demand transaction file again and perform following operations for practice
a.	Handle the missing values in the column Percent_Quantity_With_Discount with strategy = mean
b.	One hot encode the columns Mapped_Sales_Type 
c.	Label binarize the column Day
d.	Label encode the column City


```python
demand_df2=pd.read_csv("Demand.csv")
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Reference:https://dzone.com/articles/imputing-missing-data-using-sklearn-simpleimputer
imp = SimpleImputer(missing_values=np.nan,strategy='mean')

demand_df['Percent_Quantity_With_Discount']=imp.fit_transform(demand_df['Percent_Quantity_With_Discount'].values.reshape(-1,1))[:,0]
```

```python
demand_df.head()
```

```python
#b. One hot encode the columns Mapped_Sales_Type
# Reference : https://stackoverflow.com/questions/37292872/how-can-i-one-hot-encode-in-python
dummy1=pd.get_dummies(demand_df['Mapped_Sales_Type'],prefix='Mapped_Sales')
```

```python
demand_df3=pd.concat([demand_df,dummy1],axis=1)
```

```python
# we can even define a method to achieve this functionality as shared in reference link
demand_df3.drop('Mapped_Sales_Type',1)
```

```python
def dummies(df,features):
    a=pd.get_dummies(df[[features]])
    b=pd.concat([df,a],axis=1)
    b=b.drop([features],axis=1)
    return b
    
```

```python
dummies(demand_df4,'City')
```
